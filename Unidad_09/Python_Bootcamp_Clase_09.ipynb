## Y c√≥mo funcionan juntos? ü§î

Cuando combin√°s Pytest con Allure Reports y Selenium, obten√©s una poderosa herramienta para automatizar pruebas web y generar informes detallados. Pytest se encarga de ejecutar las pruebas, Selenium interact√∫a con la aplicaci√≥n web, y Allure Reports genera informes visuales de los resultados.

Ac√° hay un flujo b√°sico de c√≥mo funcionan juntos:
1. **Escrib√≠s pruebas con Pytest:** Us√°s Pytest para definir tus pruebas automatizadas. Estas pruebas pueden incluir interacciones con una aplicaci√≥n web utilizando Selenium.
2. **Ejecut√°s las pruebas:** Cuando ejecut√°s tus pruebas con Pytest, este framework maneja la ejecuci√≥n y registra los resultados.
3. **Selenium interact√∫a con la web:** Durante la ejecuci√≥n de las pruebas, Selenium se encarga de abrir el navegador, navegar por la aplicaci√≥n web, y realizar las acciones definidas en tus pruebas (como hacer clic en botones, completar formularios, etc.).
4. **Gener√°s informes con Allure Reports:** Despu√©s de que las pruebas se ejecutan, Allure Reports toma los resultados registrados por Pytest y genera informes visuales detallados. Estos informes incluyen informaci√≥n sobre qu√© pruebas pasaron, cu√°les fallaron, capturas de pantalla, logs, y m√°s.
5. **Analiz√°s los resultados:** Finalmente, pod√©s revisar los informes generados por Allure Reports para entender el estado de tu aplicaci√≥n web y detectar cualquier problema que necesite ser abordado.

### Allure Steps

Allure Steps es una caracter√≠stica de Allure Reports que permite dividir las pruebas en pasos l√≥gicos y documentar cada uno de ellos. Esto ayuda a mejorar la legibilidad de los informes y facilita la identificaci√≥n de problemas espec√≠ficos en el flujo de la prueba.

Los steps (pasos) permiten que el reporte muestre el test como una "historia".  En vez de tener un solo bloque de informaci√≥n, pod√©s ver cada paso que se ejecut√≥, lo que facilita entender qu√© hizo el test y d√≥nde pudo haber fallado. Es decir:

- Paso 1: Abrir login.
- Paso 2: Ingresar credenciales.
- Paso 3: Validar Productos.

Ejemplo B√°sico:

```python
import pytest
import allure

def test_login_ok(driver):
    with allure.step("Abrir SauceDemo"):
        driver.get("https://www.saucedemo.com/")

    with allure.step("Validar t√≠tulo"):
        assert "Swag Labs" in driver.title
```

## Screenshots en Allure üì∏

El screenshot es la evidencia N¬∫1 en UI testing.
La idea es tomar capturas de pantalla en momentos clave de la ejecuci√≥n del test, como cuando ocurre un error o al finalizar una acci√≥n importante. Estas capturas se adjuntan autom√°ticamente al reporte de Allure, proporcionando una referencia visual que ayuda a entender el estado de la aplicaci√≥n en esos momentos espec√≠ficos.

Podemos definir una funci√≥n para tomar screenshots y adjuntarlos al reporte de Allure:


## Asserts ‚úîÔ∏è

Las assertions son fundamentales en cualquier framework de testing, y Pytest no es la excepci√≥n. Una assertion es una declaraci√≥n que verifica si una condici√≥n es verdadera. Si la condici√≥n es falsa, la prueba falla.

Qu√© diferencia hay entre usar `assert` y otras formas de validaci√≥n? Bueno, `assert` es una forma simple y directa de validar condiciones en Python. Pytest mejora la experiencia de las assertions al proporcionar mensajes de error detallados autom√°ticamente cuando una assertion falla. Si ejecutas un assert mediante Pytest, la consola te mostrar√° exactamente qu√© sali√≥ mal, lo que facilita la depuraci√≥n. En cambio, si us√°s otras formas de validaci√≥n, como `if` statements con excepciones personalizadas, ten√©s que manejar los mensajes de error vos mismo, lo que puede ser m√°s tedioso y menos informativo.

Algunos ejemplos comunes de asserts:

````python
import pytest

assert 5 > 3, "5 deber√≠a ser mayor que 3."
assert "python" in "python bootcamp", "La palabra 'python' deber√≠a estar en 'python bootcamp'."
assert len([1, 2, 3]) == 3, "La longitud de la lista deber√≠a ser 3."
````

Comparaciones:

````python
import pytest

actual = True
esperado = True
assert actual == esperado, "El valor actual deber√≠a ser igual al esperado."
esperado = False
assert actual != esperado, "El valor actual deber√≠a ser diferente al esperado."
assert actual > 0, "El valor actual deber√≠a ser mayor que cero."
````

Como vemos ac√°, Pytest muestra mensajes claros cuando una aserci√≥n falla, lo que facilita la identificaci√≥n del problema.

El error es legible y el error ya no es un mensaje "crudo" de Python, sino un mensaje claro que indica qu√© condici√≥n no se cumpli√≥.

Es importante destacar que Pytest ofrece una amplia gama de funcionalidades adicionales para trabajar con assertions, como la capacidad de agrupar m√∫ltiples assertions, manejar excepciones esperadas y m√°s. Esto hace que Pytest sea una herramienta poderosa y flexible para la validaci√≥n de pruebas en Python.

Por qu√© actual es mayor que 0? Porque en Python, los valores booleanos `True` y `False` se comportan como los enteros `1` y `0`, respectivamente. Entonces, cuando hacemos la comparaci√≥n `actual > 0`, estamos verificando si `True` (que es equivalente a `1`) es mayor que `0`, lo cual es cierto.

Veamos un ejemplo donde un assert falla:

````python
import pytest

def test_fallido():
    actual = True
    esperado = False
    assert actual == esperado, "El valor actual deber√≠a ser igual al esperado."

test_fallido()
````

Aqu√≠ se ve claramente c√≥mo Pytest informa el fallo de la aserci√≥n, mostrando el mensaje personalizado que definimos.

## Marcadores üè∑Ô∏è

Los marcadores permiten clasificar, agrupar y ejecutar selectivamente tests.
Son fundamentales en proyectos reales donde no todos los tests se ejecutan siempre.

En automatizaci√≥n profesional **no se corre todo siempre**.

Un marcador es una etiqueta que se le asigna a un test para indicar:

- Qu√© tipo de test es
- Cu√°ndo deber√≠a ejecutarse
- A qu√© grupo pertenece

Ejemplos t√≠picos:
- smoke
- regression
- critical
- slow
- ui
- api
o bien, el nombre de una funcionalidad espec√≠fica del test.
- login
- consulta_saldos
- etc.

### Formas de Ejecuci√≥n

Ejemplo simple:


Aqu√≠, hemos marcado el test `test_login_exitoso` con el marcador `login`. Es una forma de categorizar este test como relacionado con la funcionalidad de login. Como si "lo bautiz√°ramos".

## Integrando Selenium + Pytest üîå

En la Unidad 8 aprendimos Selenium como ‚Äúautomatizaci√≥n de navegador‚Äù.
Ahora, el objetivo es que Selenium deje de ser scripts sueltos y pase a ser parte de una **suite de tests**:

- repetible.
- mantenible.
- ejecutable por lotes.
- con setup/teardown consistente.

### Configurar Selenium como fixture

La integraci√≥n correcta empieza por una regla:

> **El test NO debe crear el driver.**  
> El driver debe proveerse por fixture.

Esto evita duplicaci√≥n, mejora el orden del proyecto, y hace que todos los tests arranquen bajo condiciones consistentes.

### Ejemplo base (driver por test)

Archivo recomendado: `tests/conftest.py`

**¬øQu√© es conftest.py?**

Es un archivo especial en Pytest que permite compartir fixtures y configuraciones entre m√∫ltiples archivos de test. Pytest lo detecta autom√°ticamente y lo usa para cargar fixtures globales. Se debe ubicar en la ra√≠z del proyecto o en la carpeta de tests.

```python
import pytest
from selenium import webdriver

@pytest.fixture
def driver():
    driver = webdriver.Chrome()
    driver.maximize_window()
    yield driver
    driver.quit()
```

**¬øQu√© logra esto?**
- Cada test recibe un navegador nuevo.
- Al terminar el test, se cierra siempre.
- El test se enfoca en validar, no en preparar el entorno.

**Ejemplo de uso en un test:**

Precondici√≥n: Tener el archivo `tests/conftest.py` con el fixture `driver` definido.

```python
import pytest

def test_open_saucedemo(driver):
    driver.get("https://www.saucedemo.com/")
    assert "Swag Labs" in driver.title
    print("P√°gina cargada correctamente.")
```

> Pytest inyecta autom√°ticamente el fixture `driver` en el test `test_open_saucedemo`.

Al ejecutar este test, Pytest utilizar√° el fixture `driver` definido en `conftest.py`, creando una instancia de Chrome, maximizando la ventana, y cerr√°ndola al finalizar el test. Esto asegura que cada test tenga un entorno limpio y consistente para ejecutar las pruebas de Selenium.

## Fixtures

Los **fixtures** son uno de los conceptos m√°s importantes de Pytest.  
Permiten preparar y limpiar el contexto de ejecuci√≥n de los tests (setup/teardown) de manera **reutilizable, limpia y profesional**.

Son funciones especiales que se utilizan para configurar y limpiar el entorno de prueba antes y despu√©s de la ejecuci√≥n de los tests. Son ideales para manejar tareas repetitivas como la inicializaci√≥n de objetos, la configuraci√≥n de bases de datos, o en nuestro caso, la gesti√≥n del navegador con Selenium.

En automatizaci√≥n con Selenium, un fixture suele ser el lugar ideal para:
- crear y configurar el `driver`
- inicializar sesi√≥n (login)
- preparar datos
- limpiar estado (logout, cerrar navegador)
- capturar evidencias ante fallas

Un fixture es una funci√≥n decorada con `@pytest.fixture` que Pytest ejecuta **antes** del test para proveer un recurso, y opcionalmente ejecuta limpieza **despu√©s** del test.

> En vez de repetir ‚Äúsetup/teardown‚Äù en cada test, lo centralizamos en fixtures.

#### Ejemplo simple (sin Selenium)

```python
import pytest

@pytest.fixture
def numeros():
    return [1, 2, 3]

def test_longitud_lista(numeros):
    assert len(numeros) == 3
```

### Alcance de los Fixtures

Por defecto, los fixtures tienen un alcance de funci√≥n, es decir, se ejecutan antes y despu√©s de cada test que los utiliza. Sin embargo, Pytest permite definir otros alcances para los fixtures, lo que puede ser √∫til en diferentes escenarios.

El scope define cu√°ntas veces Pytest ejecuta el fixture:

| Scope | Se ejecuta... | Uso t√≠pico |
|-------|----------------|------------|
| function | antes y despu√©s de cada test que lo usa | m√°ximo aislamiento |
| class | una vez por clase | suite agrupada |
| module | una vez por m√≥dulo (archivo) | tests relacionados |
| session | una vez por sesi√≥n de Pytest (corrida) | configuraci√≥n global |

Ejemplo:

```python
import pytest

@pytest.fixture(scope="function")
def datos():
    return {"env": "test"}
```

`function` es el valor por defecto y es el m√°s seguro: cada test arranca "limpio".


### Fixtures reutilizables

Un fixture reutilizable debe:

- Tener una sola responsabilidad.
- Ser f√°cil de entender.
- Servir para muchos tests.
- Evitar hardcodear l√≥gica espec√≠fica de un test.

**Antipatr√≥n (malo): fixture con demasiadas responsabilidades**

```python
import pytest

@pytest.fixture
def driver_y_login_y_carrito():
    ...
```

Este fixture hace demasiado: crea el driver, hace login, y prepara un carrito de compras. Si un test solo necesita el driver, est√° forzado a cargar toda esa l√≥gica innecesaria.

**Buen patr√≥n (bueno): fixture con responsabilidad √∫nica**

```python
import pytest

@pytest.fixture
def driver():
    ...

@pytest.fixture
def login(driver):
    ...
```

Cada fixture tiene una responsabilidad clara y puede ser combinado seg√∫n las necesidades de cada test.


### Fixtures para setup/teardown

En automatizaci√≥n, es normal necesitar acciones repetidas antes y despu√©s de cada test, como abrir y cerrar un navegador, loguearse, volver a un estado base, etc. Los fixtures nos permiten hacer esto de manera ordenada y reutilizable. Y en Pytest, los fixtures son f√°ciles de definir y usar para estos casos.

### Fixture de Espera Expl√≠cita

```python
import pytest
from selenium.webdriver.support.ui import WebDriverWait

@pytest.fixture
def wait(driver):
    return WebDriverWait(driver, 10)
```

Este fixture `wait` proporciona una instancia de `WebDriverWait` que puede ser utilizada en los tests para implementar esperas expl√≠citas, mejorando la estabilidad de las pruebas al esperar condiciones espec√≠ficas antes de continuar con las acciones.

### Fixture para ir a un Login

```python
import pytest

@pytest.fixture
def go_to_login(driver):
    driver.get("https://www.saucedemo.com/")
    return driver
```

Este fixture `go_to_login` navega a la p√°gina de login de la aplicaci√≥n antes de que se ejecute cualquier test que lo utilice. Esto asegura que todos los tests que dependan de la p√°gina de login comiencen desde el mismo punto.

### Fixture de Login Ok

```python
import pytest

@pytest.fixture
def login_ok(driver):
    driver.get("https://www.saucedemo.com/")
    driver.find_element("id", "user-name").send_keys("standard_user")
    driver.find_element("id", "password").send_keys("secret_sauce")
    driver.find_element("id", "login-button").click()

    # Validaci√≥n m√≠nima de que entr√≥
    wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "title")))

    return driver
```

Este fixture `login_ok` realiza un login exitoso en la aplicaci√≥n. Navega a la p√°gina de login, completa las credenciales y hace clic en el bot√≥n de login. Adem√°s, incluye una validaci√≥n m√≠nima para asegurarse de que el login fue exitoso antes de devolver el driver para su uso en los tests.

### Modo de uso en tests

```python
def test_login_page_title(go_to_login):
    assert "Swag Labs" in go_to_login.title

def test_login_success(login_ok):
    assert "Products" in login_ok.title
```

Aqu√≠, los tests `test_login_page_title` y `test_login_success` utilizan los fixtures `go_to_login` y `login_ok`, respectivamente. Esto permite que cada test comience en el estado adecuado sin necesidad de repetir el c√≥digo de configuraci√≥n dentro de cada test.

Esto ya va tomando forma de un proyecto real.

## Page Object Model (POM) üì¶

El **Page Object Model (POM)** es un patr√≥n de dise√±o que propone una regla simple:

> **Cada p√°gina de la aplicaci√≥n se representa como una clase.**  
> Esa clase encapsula la forma de interactuar con la p√°gina.

El archivo de test **no deber√≠a conocer**:
- selectores.
- detalles del HTML.
- c√≥mo se hace click.
- c√≥mo se escribe en inputs.

Eso es responsabilidad del *Page Object*.

Recondando, la estructura era:

```bash
pages/
 ‚îú‚îÄ‚îÄ login_page.py
 ‚îî‚îÄ‚îÄ products_page.py

tests/
 ‚îî‚îÄ‚îÄ test_login.py
```

Los selectores y la l√≥gica de interacci√≥n con la p√°gina van en el Page Object.  Los tests s√≥lo llaman a m√©todos del Page Object, y en los test solamente habr√°n asserts y l√≥gica de validaci√≥n.

### Page Object de ejemplo: Login Page

Archivo `pages/po_login.py`

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

class PoLogin:

    def __init__(self, driver, wait):
        self.driver = driver
        self.wait = wait

    def open(self, url):
        self.driver.get(url)

    def login(self, username, password):
        self.wait.until(EC.visibility_of_element_located((By.ID, "user-name"))).send_keys(username)
        self.driver.find_element(By.ID, "password").send_keys(password)
        self.driver.find_element(By.ID, "login-button").click()
```

Archivo `tests/test_login.py`

```python
def test_login_ok(driver, wait):
    login_page = LoginPage(driver, wait)
    login_page.open("https://www.saucedemo.com/")
    login_page.login("standard_user", "secret_sauce")

    assert "inventory" in driver.current_url
```

Como podemos ver, el test `test_login_ok` no conoce los selectores ni la forma de interactuar con la p√°gina de login. Simplemente utiliza el Page Object `LoginPage` para abrir la p√°gina y realizar el login. Esto hace que los tests sean m√°s limpios, f√°ciles de leer y mantener.
En cambio, la clase `PoLogin` encapsula toda la l√≥gica de interacci√≥n con la p√°gina de login, incluyendo los selectores y las acciones necesarias para completar el proceso de login.
