Seg√∫n el paper ["Comparison of JSON and XML Data Formats"](https://www.researchgate.net/publication/329707959_Comparison_of_JSON_and_XML_Data_Formats) de [Dr. Alen ≈†imec](https://www.researchgate.net/profile/Alen-Simec), publicado en 2014, JSON no reemplazar√° por completo a XML, sino que ambos formatos coexistir√°n: JSON como la opci√≥n m√°s pr√°ctica para flujos de datos simples y de alto rendimiento en entornos web, y XML como la herramienta de elecci√≥n cuando se requiere un est√°ndar robusto, amplio en funciones y con validaci√≥n estricta de documentos.

### XML  üìÑ

XML (eXtensible Markup Language) es un lenguaje de marcado que se utiliza para almacenar y transportar datos. Fue dise√±ado para ser legible tanto por humanos como por m√°quinas, y es ampliamente utilizado en aplicaciones web y servicios web debido a su flexibilidad y capacidad para representar datos estructurados.

### Caracter√≠sticas Principales

* **Estructura Jer√°rquica**: XML organiza los datos en una estructura de √°rbol, donde los elementos pueden contener otros elementos, creando una jerarqu√≠a clara y f√°cil de entender.
* **Etiquetas Personalizadas**: A diferencia de HTML, XML permite a los usuarios definir sus propias etiquetas, lo que proporciona una gran flexibilidad para representar diferentes tipos de datos.
* **Legibilidad**: XML est√° dise√±ado para ser legible tanto por humanos como por m√°quinas, lo que facilita la comprensi√≥n y el procesamiento de los datos.
* **Independencia de Plataforma**: XML es independiente de la plataforma y del lenguaje de programaci√≥n, lo que permite su uso en una amplia variedad de aplicaciones y sistemas.
* **Validaci√≥n**: XML puede ser validado utilizando DTD (Document Type Definition) o XML Schema, lo que garantiza que los datos cumplen con una estructura espec√≠fica y son consistentes.
* **Soporte para Unicode**: XML soporta Unicode, lo que permite representar caracteres de diferentes idiomas y conjuntos de caracteres.

### Estructura B√°sica de un Documento XML

Un documento XML t√≠pico tiene la siguiente estructura b√°sica:


### Elementos y Atributos
* **Elementos**: Son las unidades b√°sicas de un documento XML y est√°n delimitados por etiquetas de apertura y cierre. Pueden contener texto, otros elementos o ambos.
* **Atributos**: Son pares clave-valor que proporcionan informaci√≥n adicional sobre un elemento. Los atributos se definen dentro de la etiqueta de apertura de un elemento.


### Parseo de XML con Python üêç

El parseo de XML en Python se puede realizar utilizando diferentes bibliotecas, siendo `xml.etree.ElementTree` una de las m√°s utilizadas debido a su simplicidad y eficiencia. A continuaci√≥n, se muestra un ejemplo b√°sico de c√≥mo parsear un documento XML utilizando esta biblioteca.

```python
    import xml.etree.ElementTree as ET

    # Cargar y parsear el archivo XML
    tree = ET.parse('archivo.xml')
    root = tree.getroot()

    # Recorrer los elementos del XML
    for child in root:
        print(child.tag, child.attrib)
        for subchild in child:
            print('  ', subchild.tag, subchild.text)
```
En este ejemplo, se carga un archivo XML llamado `archivo.xml`, se obtiene la ra√≠z del documento y se recorre cada elemento hijo, imprimiendo su etiqueta y atributos. Luego, se recorren los subelementos de cada hijo, imprimiendo sus etiquetas y contenido de texto.

### Parseo de JSON con Python üêç

Parsear. ¬øQu√© significa parsear algo? Significa analizarlo y descomponerlo en sus partes divididas para entender su estructura y contenido. En el contexto de la programaci√≥n, parsear un documento o una cadena de texto implica convertirlo en un formato que pueda ser f√°cilmente manipulado y procesado.

El parseo de un archivo JSON en Python se puede realizar utilizando la biblioteca est√°ndar `json`, que proporciona funciones para convertir cadenas JSON en objetos de Python y viceversa.  Esta biblioteca ya viene incluida con Python, por lo que no es necesario instalar nada adicional.

A continuaci√≥n, un ejemplo b√°sico de c√≥mo parsear un documento JSON utilizando esta biblioteca.

```python

    # Cargar y parsear el archivo JSON
```

En este ejemplo , se carga un archivo JSON llamado `archivo.json`, se parsea su contenido y se accede a los datos del JSON, imprimiendo el valor asociado a la clave 'clave' para cada elemento en el JSON.

### Serializaci√≥n y Deserializaci√≥n

La serializaci√≥n es el proceso de convertir un objeto o estructura de datos en un formato que pueda ser almacenado o transmitido, como JSON o XML. La deserializaci√≥n es el proceso inverso, donde se convierte el formato almacenado o transmitido de vuelta a su forma original como un objeto o estructura de datos.

Por ejemplo:
```python
    import json

    # Serializaci√≥n: Convertir un diccionario de Python a una cadena JSON
    data = {'nombre': 'Juan', 'edad': 30, 'ciudad': 'Madrid'}
    json_string = json.dumps(data)
    print(json_string)  # Salida: {"nombre": "Juan", "edad": 30, "ciudad": "Madrid"}

    # Deserializaci√≥n: Convertir una cadena JSON de vuelta a un diccionario de Python
    json_data = '{"nombre": "Ana", "edad": 25, "ciudad": "Barcelona"}'
    data_dict = json.loads(json_data)
    print(data_dict)  # Salida: {'nombre': 'Ana', 'edad': 25, 'ciudad': 'Barcelona'}
```

Aqu√≠ vemos c√≥mo se utiliza la biblioteca `json` de Python para serializar un diccionario de Python a una cadena JSON utilizando `json.dumps()`, y luego deserializar una cadena JSON de vuelta a un diccionario de Python utilizando `json.loads()`.

El m√©todo `json.dumps()` se utiliza para escribir directamente un objeto de Python en un archivo JSON, mientras que `json.loads()` se utiliza para leer un archivo JSON y convertir su contenido en un objeto de Python.

````python
import json

# Serializaci√≥n
data = {'nombre': 'Juan', 'edad': 30, 'ciudad': 'Madrid'}  # Diccionario de Python
json_string = json.dumps(data)  # Lo convertimos a una cadena JSON
print(json_string)
print(type(json_string))  # Muestra: <class 'str'>, es una cadena JSON.
````

````python
import json

# Deserializaci√≥n
json_data = '{"nombre": "Ana", "edad": 25, "ciudad": "Barcelona"}'  # Cadena JSON
data_dict = json.loads(json_data)  # Lo convertimos a un diccionario de Python
print(data_dict)
print(type(data_dict))  # Muestra: <class 'dict'>, es un diccionario de Python.
````

## API's  üõ†Ô∏è

Las API's (Interfaces de Programaci√≥n de Aplicaciones) son un conjunto de reglas y protocolos que permiten la comunicaci√≥n entre diferentes aplicaciones y servicios. Facilitan la integraci√≥n de funcionalidades y datos de una aplicaci√≥n en otra, lo que acelera el desarrollo y mejora la interoperabilidad.

Simplifican el desarrollo de aplicaciones y software permitiendo a los desarrolladores integrar datos, servicios y capacidades de otras aplicaciones, en lugar de desarrollarlas desde cero para cada plataforma.

Es √∫til pensar en la comunicaci√≥n de la API en t√©rminos de una solicitud y respuesta entre un cliente y un servidor. La aplicaci√≥n que env√≠a la solicitud es el cliente y el servidor proporciona la respuesta. La API es el puente que establece la conexi√≥n entre ellos.

### REST API's

Una API REST *(Representational State Transfer)* es un conjunto de reglas y convenciones para crear servicios web que permiten la comunicaci√≥n entre diferentes sistemas a trav√©s de HTTP. Las API REST son ampliamente utilizadas en aplicaciones web y m√≥viles para acceder y manipular datos de manera eficiente.

#### Principios REST
Las API REST se basan en varios principios fundamentales:
* **Stateless (Sin Estado)**: Cada solicitud del cliente al servidor debe contener toda la informaci√≥n necesaria para entender y procesar la solicitud. El servidor no debe almacenar ning√∫n estado del cliente entre solicitudes.
* **Recursos**: Los recursos son los objetos o datos que la API maneja, y cada recurso se identifica mediante una URL √∫nica. Por ejemplo, en una API de usuarios, un recurso podr√≠a ser un usuario espec√≠fico identificado por su ID.
* **M√©todos HTTP**: Repasando, las API REST utilizan los m√©todos HTTP est√°ndar para realizar operaciones sobre los recursos. Los m√©todos m√°s comunes son:
  - **GET**: Recuperar datos de un recurso.
  - **POST**: Crear un nuevo recurso.
  - **PUT**: Actualizar un recurso existente.
  - **DELETE**: Eliminar un recurso.

#### Repaso: Consumo de APIs REST desde Python (con requests)

Un repaso cortito de la Unidad 5:

```python
    import requests

    # Hacer una solicitud GET a una API REST
    response = requests.get('https://api.example.com/resource')

    # Verificar el estado de la respuesta
    if response.status_code == 200:
        data = response.json()  # Parsear la respuesta JSON
        print(data)
    else:
        print(f'Error: {response.status_code}')
```

### SOAP API's

SOAP (Simple Object Access Protocol) es un protocolo de mensajer√≠a basado en XML que se utiliza para intercambiar informaci√≥n estructurada en la implementaci√≥n de servicios web. SOAP es independiente del lenguaje y la plataforma, lo que permite la comunicaci√≥n entre aplicaciones desarrolladas en diferentes entornos. 

Aunque **REST** es hoy el est√°ndar m√°s popular, SOAP todav√≠a se usa mucho en entornos empresariales y sistemas legados (por ejemplo, en banca, seguros, y gobierno).

| Caracter√≠stica         | REST                                  | SOAP                                  |
|-----------------------|--------------------------------------|--------------------------------------|
| **Tipo**             | Estilo arquitect√≥nico                 | Protocolo de mensajer√≠a              |
| **Formato de datos** | JSON (principalmente) o XML           | Exclusivamente XML                   |
| **Contrato**         | No necesita contrato formal           | Requiere un WSDL (contrato)          |
| **Facilidad**        | M√°s simple, liviano y flexible        | M√°s complejo y verboso               |
| **Casos de uso**     | APIs modernas, microservicios, apps   | Sistemas legados, transacciones cr√≠ticas, integraci√≥n empresarial |


#### Mensajes XML y contratos WSDL

SOAP utiliza mensajes XML muy estructurados. Cada mensaje tiene un formato est√°ndar que incluye:

- **Envelope**: indica que es un mensaje SOAP.
- **Header**: informaci√≥n opcional (autenticaci√≥n, transacciones).
- **Body**: los datos o la operaci√≥n que se quiere ejecutar.
- **Fault**: bloque de error si algo falla.

### Ejemplo de Mensaje SOAP

```xml
<soap:Envelope xmlns:soap="http://schemas.xmlsoap.org/soap/envelope/">
   <soap:Header/>
   <soap:Body>
      <m:GetPrice xmlns:m="https://www.ejemplo.com/stock">
         <m:Item>Notebook</m:Item>
      </m:GetPrice>
   </soap:Body>
</soap:Envelope>
```

#### SOAP Requests con Python

Para consumir un servicio SOAP en Python, podemos usar la biblioteca `requests` que vimos en la unidad anterior. Tambi√©n podemos usar la librer√≠a `lxml` para construir y parsear mensajes XML.
Existen otras biblitecas especializadas como `zeep` o `suds-jurko` que facilitan mucho el trabajo con SOAP, pero aqu√≠ haremos un ejemplo simple con `requests`, dado que es la biblioteca m√°s popular y f√°cil de usar.

```python
      import requests
   
      # Definir el endpoint del servicio SOAP
      url = 'https://www.ejemplo.com/soap'
   
      # Mensaje SOAP
      body = '''<?xml version="1.0"?>
      <soap:Envelope xmlns:soap="http://schemas.xmlsoap.org/soap/envelope/">
         <soap:Header/>
         <soap:Body>
            <m:GetPrice xmlns:m="https://www.ejemplo.com/stock">
               <m:Item>Notebook</m:Item>
            </m:GetPrice>
         </soap:Body>
      </soap:Envelope>'''
   
      # Definir los headers HTTP
      headers = {
         'Content-Type': 'text/xml; charset=utf-8',
         'SOAPAction': 'https://www.ejemplo.com/GetPrice'
      }
   
      # Hacer la solicitud POST al servicio SOAP
      response = requests.post(url, data=body, headers=headers)
   
      # Verificar el estado de la respuesta
      if response.status_code == 200:
         print('Respuesta SOAP:')
         print(response.text)  # Aqu√≠ podr√≠amos parsear el XML de la respuesta
      else:
         print(f'Error: {response.status_code}')
   ```

## Autenticaci√≥n y Autorizaci√≥n üîê

La autenticaci√≥n API verifica qui√©n realiza la solicitud; la autorizaci√≥n API determina a qu√© pueden acceder.
Los m√©todos de autenticaci√≥n comunes incluyen autenticaci√≥n b√°sica, API key, JSON web tokens (JWT), OAuth 2.0 y otras.

### Autenticaci√≥n B√°sica

La autenticaci√≥n b√°sica es un m√©todo simple y com√∫nmente utilizado para proteger las API. Funciona enviando las credenciales del usuario (nombre de usuario y contrase√±a) en cada solicitud HTTP. Aunque es f√°cil de implementar, tiene limitaciones de seguridad.

Para autenticarse, el cliente env√≠a otra solicitud al servidor con el encabezado de autorizaci√≥n incluido en el encabezado de la solicitud HTTP, que contiene la palabra "Basic" seguida de una cadena codificada en base64 con el nombre de usuario y la contrase√±a. Por ejemplo, "Authorization: Basic dXNlcm5hbWU6cGFzc3dvcmQ=" representa las credenciales "username:password" codificadas en base64.
